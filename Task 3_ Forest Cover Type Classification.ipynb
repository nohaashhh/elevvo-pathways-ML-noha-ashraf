```python
X = df.drop("Cover_Type", axis=1)
y = df["Cover_Type"] - 1
```

```python
#Train–Test Split
import numpy as np

np.random.seed(42)
indices = np.random.permutation(len(X))
train_size = int(0.8 * len(X))

train_idx = indices[:train_size]
test_idx  = indices[train_size:]

X_train = X.iloc[train_idx]
X_test  = X.iloc[test_idx]
y_train = y.iloc[train_idx]
y_test  = y.iloc[test_idx]
```

```python
pip install xgboost
```

```python
#XGBoost Data Format
import xgboost as xgb

dtrain = xgb.DMatrix(X_train, label=y_train)
dtest  = xgb.DMatrix(X_test, label=y_test)
```

```python
#XGBoost Multi-Class Model
params = {
    "objective": "multi:softmax",
    "num_class": 7,
    "max_depth": 6,
    "eta": 0.1,
    "subsample": 0.8,
    "colsample_bytree": 0.8,
    "eval_metric": "merror"
}

model = xgb.train(
    params,
    dtrain,
    num_boost_round=100
)
```

```python
#predictions
y_pred = model.predict(dtest)
```

```python
#Evaluation Metrics
accuracy = np.mean(y_pred == y_test.values)
print("Accuracy:", accuracy)
```

```python
#confusion matrix
num_classes = 7
conf_matrix = np.zeros((num_classes, num_classes), dtype=int)

for t, p in zip(y_test.values, y_pred.astype(int)):
    conf_matrix[int(t), int(p)] += 1

conf_matrix
```

```python
#visualization
import matplotlib.pyplot as plt

plt.imshow(conf_matrix, cmap="Blues")
plt.colorbar()
plt.xlabel("Predicted")
plt.ylabel("Actual")
plt.title("Confusion Matrix")
plt.show()
```

```python
#feature importance
importance = model.get_score(importance_type="gain")

imp_df = pd.DataFrame({
    "Feature": importance.keys(),
    "Importance": importance.values()
}).sort_values(by="Importance", ascending=False)

imp_df.head(10)
```

```python
#plottop 10 feature importance
plt.barh(
    imp_df["Feature"][:10][::-1],
    imp_df["Importance"][:10][::-1]
)
plt.title("Top 10 Feature Importances")
plt.show()
```

```python
#Random Forest vs XGBoost
rf_params = {
    "objective": "multi:softmax",
    "num_class": 7,
    "max_depth": 10,
    "subsample": 0.6,
    "colsample_bynode": 0.6,
    "booster": "gbtree",
    "eval_metric": "merror"
}

rf_model = xgb.train(
    rf_params,
    dtrain,
    num_boost_round=200
)

rf_pred = rf_model.predict(dtest)
rf_acc = np.mean(rf_pred == y_test.values)

print("Random Forest-style Accuracy:", rf_acc)
```

```python
#Hyperparameter Tuning
depths = [4, 6, 8]
etas = [0.05, 0.1]

best_acc = 0
best_params = None

for d in depths:
    for e in etas:
        params["max_depth"] = d
        params["eta"] = e

        model = xgb.train(params, dtrain, num_boost_round=100)
        preds = model.predict(dtest)
        acc = np.mean(preds == y_test.values)

        if acc > best_acc:
            best_acc = acc
            best_params = (d, e)

print("Best Accuracy:", best_acc)
print("Best Params (depth, eta):", best_params)
```

XGBoost was used for multi-class forest cover classification due to its efficiency and robustness.
The model achieved high accuracy by leveraging elevation and distance-based features.
Feature importance analysis showed elevation as the most influential factor.
A Random Forest–style model was also tested for comparison, while manual hyperparameter tuning improved performance
