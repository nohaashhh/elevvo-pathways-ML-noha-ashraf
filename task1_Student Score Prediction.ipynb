```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np
```

```python
import kagglehub

# Download latest version
path = kagglehub.dataset_download("lainguyn123/student-performance-factors")

print("Path to dataset files:", path)
```

```python
import os

os.listdir(path)
```

```python
# Load the dataset
df = pd.read_csv(os.path.join(path, "StudentPerformanceFactors.csv"))
df.head()
```

```python
df.info()
df.describe()
```

```python
#data cleaning
# Check missing values
df.isnull().sum()
# Drop rows with missing values
df = df.dropna()
```

```python
df.isnull().sum()
```

```python
#Visualization
#study hours vs exam score
plt.figure(figsize=(6,4))
plt.scatter(df['Hours_Studied'], df['Exam_Score'], color='blue')
plt.xlabel("Study Hours")
plt.ylabel("Exam Score")
plt.title("Study Hours vs Exam Score")
plt.show()
```

```python
# Convert categorical variables to numerical using one-hot encoding
df = pd.get_dummies(df, drop_first=True)
df.head()
```

```python
#correlation heatmap
plt.figure(figsize=(10,6))
sns.heatmap(df.corr(), annot=True, cmap="coolwarm")
plt.show()
```

```python
#feature selection
X = df[['Hours_Studied']]
y = df['Exam_Score']
```

```python
df_shuffled = df.sample(frac=1, random_state=42)

train_size = int(0.8 * len(df_shuffled))

train_df = df_shuffled[:train_size]
test_df  = df_shuffled[train_size:]

X_train = train_df[['Hours_Studied']]
y_train = train_df['Exam_Score']

X_test = test_df[['Hours_Studied']]
y_test = test_df['Exam_Score']
```

```python
#linear regression model
# Add bias term
X_train_np = np.c_[np.ones(len(X_train)), X_train.values]
X_test_np  = np.c_[np.ones(len(X_test)), X_test.values]

y_train_np = y_train.values
y_test_np  = y_test.values

# Normal Equation
theta = np.linalg.inv(X_train_np.T @ X_train_np) @ X_train_np.T @ y_train_np

# Predictions
y_pred = X_test_np @ theta
# Evaluate the model
mse = np.mean((y_test_np - y_pred) ** 2)
print(f"Mean Squared Error: {mse}")
```

```python
#plotting the regression line
plt.figure(figsize=(6,4))
plt.scatter(X_test, y_test, color='blue', label='Actual Data')
plt.plot(X_test, y_pred, color='red', label='Regression Line')
plt.xlabel("Study Hours")
plt.ylabel("Exam Score")
plt.title("Linear Regression: Study Hours vs Exam Score")
plt.legend()
plt.show()
```

```python
#accuracy
accuracy = 1 - (np.sum(np.abs(y_test_np - y_pred)) / np.sum(y_test_np))
print(f"Model Accuracy: {accuracy * 100:.2f}%")
```

```python
#Evaluation
import numpy as np

y_true = y_test.values
y_pred = y_pred  # already NumPy array

mae = np.mean(np.abs(y_true - y_pred))
print("MAE:", mae)

#mse score
rmse = np.sqrt(np.mean((y_true - y_pred) ** 2))
print("RMSE:", rmse)


#R^2 Score
ss_res = np.sum((y_true - y_pred) ** 2)
ss_tot = np.sum((y_true - np.mean(y_true)) ** 2)

r2 = 1 - (ss_res / ss_tot)
print("R:", r2)

```

```python
#trying polynomial regression

features = ['Hours_Studied']
X = df[features]
y = df['Exam_Score']
```

```python
#train test split
import numpy as np

np.random.seed(42)
indices = np.random.permutation(len(X))
train_size = int(0.8 * len(X))

train_idx = indices[:train_size]
test_idx  = indices[train_size:]

X_train = X.iloc[train_idx]
X_test  = X.iloc[test_idx]
y_train = y.iloc[train_idx]
y_test  = y.iloc[test_idx]
```

```python
#create polynomial features =2
# Convert to NumPy
X_train_np = X_train.values
X_test_np  = X_test.values

# Polynomial features: [1, x, x]
X_train_poly = np.c_[np.ones(len(X_train_np)),
                     X_train_np,
                     X_train_np ** 2]

X_test_poly = np.c_[np.ones(len(X_test_np)),
                    X_test_np,
                    X_test_np ** 2]
```

```python
#Train Using Normal Equation
y_train_np = y_train.values
y_test_np  = y_test.values

theta = np.linalg.inv(
    X_train_poly.T @ X_train_poly
) @ X_train_poly.T @ y_train_np
```

```python
#predictions
y_pred_poly = X_test_poly @ theta
```

```python
#evaluation
mae = np.mean(np.abs(y_test_np - y_pred_poly))
mse = np.mean((y_test_np - y_pred_poly) ** 2)

ss_res = np.sum((y_test_np - y_pred_poly) ** 2)
ss_tot = np.sum((y_test_np - np.mean(y_test_np)) ** 2)
r2 = 1 - ss_res / ss_tot

print("Polynomial MAE:", mae)
print("Polynomial MSE:", mse)
print("Polynomial R:", r2)
```

```python
#Feature Combination Experiments
features = ['Hours_Studied', 'Attendance', 'Previous_Scores']
X = df[features]

y = df['Exam_Score']

```

```python
#Training
np.random.seed(42)
indices = np.random.permutation(len(X))
train_size = int(0.8 * len(X))
train_idx = indices[:train_size]
test_idx  = indices[train_size:]
X_train = X.iloc[train_idx]
X_test  = X.iloc[test_idx]
y_train = y.iloc[train_idx]
y_test  = y.iloc[test_idx]
```

```python
# Add bias
X_train_b = np.c_[np.ones(len(X_train)), X_train]
X_test_b  = np.c_[np.ones(len(X_test)), X_test]

# Train
theta = np.linalg.inv(
    X_train_b.T @ X_train_b
) @ X_train_b.T @ y_train

# Predict
y_pred = X_test_b @ theta
```

```python
ss_res = np.sum((y_test - y_pred) ** 2)
ss_tot = np.sum((y_test - np.mean(y_test)) ** 2)
r2 = 1 - ss_res / ss_tot

print("Features:", features)
print("R:", r2)
```

```python
#accuracy
accuracy = 1 - (np.sum(np.abs(y_test.values - y_pred)) / np.sum(y_test.values))
print(f"Model Accuracy with features {features}: {accuracy * 100:.2f}%")
```

```python
#visualization
plt.figure(figsize=(10,6))      
sns.barplot(x=features, y=theta[1:])
plt.title("Feature Coefficients")
plt.ylabel("Coefficient Value")
plt.show()
```

```python
#plotting the regression line
plt.figure(figsize=(6,4))
plt.scatter(X_test['Hours_Studied'], y_test, color='blue', label='Actual Data')
plt.scatter(X_test['Hours_Studied'], y_pred, color='red', label='Predicted Data')
plt.xlabel("Study Hours")
plt.ylabel("Exam Score")
plt.title("Multiple Linear Regression: Study Hours vs Exam Score")
plt.legend()
plt.show()
```

Polynomial regression slightly improved performance, indicating a non-linear relationship between study hours and exam scores.
Adding features such as sleep hours and attendance further increased R, showing that academic performance depends on multiple behavioral factors rather than study time alone.
