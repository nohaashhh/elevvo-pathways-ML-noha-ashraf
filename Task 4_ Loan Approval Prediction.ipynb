```python
#Encode Categorical Features

categorical_cols = df.select_dtypes(include=['object']).columns
categorical_cols

```

```python
from collections import defaultdict

label_maps = defaultdict(dict)

for col in df.select_dtypes(include='object').columns:
    unique_vals = df[col].unique()
    label_maps[col] = {v: i for i, v in enumerate(unique_vals)}
    df[col] = df[col].map(label_maps[col])
```

```python
#split features and target variable
X = df.drop(' loan_status', axis=1)
y = df[' loan_status']
X
```

```python
#Handle Class Imbalance
approved = df[df[" loan_status"] == 1]
rejected = df[df[" loan_status"] == 0]

# Oversample minority class
if len(approved) < len(rejected):
    approved = approved.sample(len(rejected), replace=True)
else:
    rejected = rejected.sample(len(approved), replace=True)

balanced_df = pd.concat([approved, rejected]).sample(frac=1)

X = balanced_df.drop(" loan_status", axis=1).values
y = balanced_df[" loan_status"].values
```

```python
#Train-Test Split (Manual)
split_index = int(0.8 * len(X))
X_train, X_test = X[:split_index], X[split_index:]
y_train, y_test = y[:split_index], y[split_index:]
print("Training samples:", len(X_train))
print("Testing samples:", len(X_test))
```

```python
#Logistic Regression 
class LogisticRegressionScratch:
    def __init__(self, lr=0.01, epochs=3000):
        self.lr = lr
        self.epochs = epochs

    def sigmoid(self, z):
        return 1 / (1 + np.exp(-z))

    def fit(self, X, y):
        self.w = np.zeros(X.shape[1])
        self.b = 0

        for _ in range(self.epochs):
            linear = np.dot(X, self.w) + self.b
            y_pred = self.sigmoid(linear)

            dw = np.dot(X.T, (y_pred - y)) / len(y)
            db = np.mean(y_pred - y)

            self.w -= self.lr * dw
            self.b -= self.lr * db

    def predict(self, X):
        return (self.sigmoid(np.dot(X, self.w) + self.b) >= 0.5).astype(int)
```

```python
#Evaluation Metrics (Precision, Recall, F1) Score
def precision_recall_f1(y_true, y_pred):
    tp = np.sum((y_true == 1) & (y_pred == 1))
    fp = np.sum((y_true == 0) & (y_pred == 1))
    fn = np.sum((y_true == 1) & (y_pred == 0))

    precision = tp / (tp + fp) if (tp + fp) > 0 else 0
    recall = tp / (tp + fn) if (tp + fn) > 0 else 0
    f1 = 2 * (precision * recall) / (precision + recall) if (precision + recall) > 0 else 0

    return precision, recall, f1
```

```python
#test train the model
log_reg = LogisticRegressionScratch()
log_reg.fit(X_train, y_train)

y_pred_lr = log_reg.predict(X_test)
p, r, f1 = precision_recall_f1(y_test, y_pred_lr)

print("Logistic Regression")
print("Precision:", round(p, 3))
print("Recall:", round(r, 3))
print("F1-score:", round(f1, 3))
```

```python
#bonus Simple Decision Tree
class SimpleDecisionTree:
    def fit(self, X, y):
        self.feature = np.argmax(np.var(X, axis=0))
        self.threshold = np.median(X[:, self.feature])

    def predict(self, X):
        return (X[:, self.feature] > self.threshold).astype(int)
```

```python
tree = SimpleDecisionTree()
tree.fit(X_train, y_train)

y_pred_tree = tree.predict(X_test)
p, r, f1 = precision_recall_f1(y_test, y_pred_tree)

print("\nDecision Tree")
print("Precision:", round(p, 3))
print("Recall:", round(r, 3))
print("F1-score:", round(f1, 3))
```
